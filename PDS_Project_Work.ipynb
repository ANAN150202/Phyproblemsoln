```python
#importing libary function
import pandas as pd
import numpy as np
```

```python
#reading and showing csv file
df= pd.read_csv("/content/survey lung cancer.csv")

descriptions = """\
Gender: M (male), F (female)
Age: Age of the patient
Smoking: YES = 2, NO = 1
Yellow fingers: YES = 2, NO = 1
Anxiety: YES = 2, NO = 1
Peer pressure: YES = 2, NO = 1
Chronic Disease: YES = 2, NO = 1
Fatigue: YES = 2, NO = 1
Allergy: YES = 2, NO = 1
Wheezing: YES = 2, NO = 1
Alcohol: YES = 2, NO = 1
Coughing: YES = 2, NO = 1
Shortness of Breath: YES = 2, NO = 1
Swallowing Difficulty: YES = 2, NO = 1
Chest pain: YES = 2, NO = 1
Lung Cancer: YES, NO
"""


print(descriptions)
df


```

```python
#Changing String to Numaric data
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['LUNG_CANCER'] = label_encoder.fit_transform(df['LUNG_CANCER'])
df['GENDER'] = label_encoder.fit_transform(df['GENDER'])
df
```

```python
#spliting data into 2 parts
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
#spliting df into 4 catagories
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train
```

```python
x_test
```

```python
print(y_test)
print(y_train)
```

```python
#model training
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=2222, solver='saga')
model.fit(x_train,y_train)
```

```python
#prediction
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
#acuracy test
from sklearn.metrics import accuracy_score
```

```python
# accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
print('Accuracy on Test data : ', test_data_accuracy)
```

```python
#visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Plot bar charts for each binary feature with respect to lung cancer
plt.figure(figsize=(15, 12))
for i, column in enumerate(df.columns[:-1], 1):  # Exclude the target column
    plt.subplot(4, 4, i)
    sns.countplot(x=column, hue="LUNG_CANCER", data=df)
    plt.title(f"{column} vs Lung Cancer")
plt.tight_layout()
plt.show()
```
